## Modelos de Séries Temporais

É uma seguencia de números que são ordenados por um eixo temporal

### Parte de uma serie temporal

1. Componente de tendencia
É um movimento  direcionadl consistente em uma série temporal. Essa tendencias serão deterministicas ou estocaticas.

2. Sazonal
Sazonal é uma componente de curto prazo que se repete em padrões regulares em uma série temporal. Exemplos de padrões sazonais incluem um padrão diário, semanal, mensal, trimestral ou anual.

Podemos escrever os componentes de uma serie temporal como $y_t$ como
$$
y_t = S_t+T_t+R_t
$$
Para decompor a serie temporal no python

```python
import statsmodels.api as sm
# dataset passengers
from  statsmodels.datasets import co2
import matplotlib.pyplot as plt
```

```python
co2data = co2.load_pandas()
```

```python
co2data.data.plot()
```

```python
series = co2data.data['co2']
```

```python
# Forward pass mean
series.ffill(inplace=True)
```

```python
decompo = sm.tsa.seasonal_decompose(co2data.data['co2'].fillna(0), model='additive')
```

```python
decompo.plot()
plt.show()
```

* Estacionaridade: quando a media e variancia da serie são constante ao lingo do tempo
> Ter tendencia ou sazonalizade = não estacionaria

Para Solucionar isso tornamos a series não esatcionaria em estacionaria, **por meio da diferenciação**

```python
serie_diff = series.diff()
serie_diff.plot()
```

```python
decompo = sm.tsa.seasonal_decompose(serie_diff.iloc[1:], model='additive', period=52)
decompo.plot()
plt.show()
```
